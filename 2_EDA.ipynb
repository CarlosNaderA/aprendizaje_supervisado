# 2. Análisis Exploratorio de Datos - Titanic

En este notebook se analizan las variables principales para entender qué factores pudieron influir en la supervivencia.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv("data/titanic_clean.csv")
df.head()
```

## Distribución de supervivencia

Primero reviso cuántas personas sobrevivieron y cuántas no.

```python
sns.countplot(data=df, x="Survived")
plt.title("Distribución de supervivencia")
plt.xlabel("Sobrevivió")
plt.ylabel("Total")
plt.show()
```

## Supervivencia por clase

La clase del boleto parece importante porque los pasajeros de primera clase tuvieron mayor acceso a zonas seguras y botes salvavidas.

```python
sns.countplot(data=df, x="Pclass", hue="Survived")
plt.title("Tasa de Supervivencia por Clase")
plt.xlabel("Clase")
plt.ylabel("Total")
plt.legend(title="Survived", labels=["No sobrevivió", "Sobrevivió"])
plt.show()
```

## Supervivencia por género

El género es una variable relevante porque en el contexto histórico del Titanic se dio prioridad a mujeres y niños.

```python
sns.countplot(data=df, x="Sex", hue="Survived")
plt.title("Supervivencia por Género")
plt.xlabel("Género")
plt.ylabel("Total")
plt.legend(title="Survived", labels=["No sobrevivió", "Sobrevivió"])
plt.show()
```

## Distribución de edad

La edad permite observar si los niños o personas jóvenes tuvieron una probabilidad distinta de supervivencia.

```python
sns.histplot(data=df, x="Age", hue="Survived", kde=True, bins=30)
plt.title("Distribución de edad según supervivencia")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

## Conclusión del EDA

Las variables más relevantes parecen ser Sex, Pclass y Age. El análisis muestra que las mujeres, los niños y los pasajeros de clases más altas tuvieron mayores probabilidades de sobrevivir.
