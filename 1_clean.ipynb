# 1. Limpieza de datos - Titanic

Este notebook realiza una limpieza básica del conjunto de datos del Titanic para preparar la información antes del análisis y del entrenamiento de modelos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
# Cargar dataset
# Si el archivo está dentro de la carpeta data del repositorio, este código funcionará directamente.
df = pd.read_csv("data/train.csv")

df.head()
```

## Revisión inicial

Primero reviso el tamaño del dataset, los tipos de datos y los valores nulos para identificar columnas que necesitan limpieza.

```python
print("Tamaño del dataset:", df.shape)
display(df.info())
display(df.isnull().sum())
```

## Tratamiento de valores nulos

La columna Age se rellena con la mediana porque es una medida resistente a valores extremos. La columna Embarked se rellena con la moda porque contiene categorías y tiene pocos valores faltantes.

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Cabin tiene demasiados valores nulos, por eso se elimina.
if "Cabin" in df.columns:
    df = df.drop(columns=["Cabin"])

df.isnull().sum()
```

## Revisión de duplicados

También verifico si existen filas repetidas en el conjunto de datos.

```python
duplicados = df.duplicated().sum()
print("Filas duplicadas:", duplicados)
```

## Eliminación de variables poco útiles

PassengerId es solo un identificador, Name es texto difícil de usar directamente y Ticket tiene muchos valores únicos. Para un modelo básico, estas columnas pueden eliminarse.

```python
columnas_eliminar = ["PassengerId", "Name", "Ticket"]
df_clean = df.drop(columns=[col for col in columnas_eliminar if col in df.columns])

df_clean.head()
```

## Guardar dataset limpio

Guardo el archivo limpio para usarlo en los siguientes notebooks.

```python
df_clean.to_csv("data/titanic_clean.csv", index=False)
print("Archivo guardado como data/titanic_clean.csv")
```
